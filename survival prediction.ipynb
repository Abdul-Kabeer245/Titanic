```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'train_data.csv')
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0', 'PassengerId'], inplace = True)
```

```python
df
```

```python
sns.countplot(data = df, x = 'Survived', hue = 'Sex')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x_train = df.drop('Survived', axis = 1)
y_train = df['Survived']
lor = LogisticRegression()
lor.fit(x_train, y_train)
```

```python
test_data = pd.read_csv(r'test_data.csv')
test_data
```

```python
test_data.drop(['Unnamed: 0', 'PassengerId'], axis = 1, inplace = True)
x_test = test_data.drop('Survived', axis = 1)
y_test = test_data['Survived']
```

```python
predict = lor.predict(x_test)
```

```python
lor.predict_proba(x_test)
```

```python
lor.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test, predict), columns = ['Predicted No', 'Predicted Yes'], index = ['Actual No', 'Actual Yes'])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predict))
```

```python
x = np.arange(len(predict))
sns.regplot(x = x, y =predict,logistic = True, ci = None)
```
